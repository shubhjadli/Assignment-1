```python
import pandas as pd
import numpy as np
data=pd.read_csv(r'/Users/shubhamjadli/Desktop/Term 3/MLP/DataSets/CarPrice_Assignment.csv')
#Seeing if there are any NA values
a=data.isna().sum()
Total_NA_Value_Columns=0
for x in a:
    if a[x]==0:
        Total_NA_Value_Columns=Total_NA_Value_Columns
    else:
        Total_NA_Value_Columns=Total_NA_Value_Columns+1
print('Total columns with NA values =',Total_NA_Value_Columns)
#finding duplicates 
print('Number of duplicates =', data.duplicated().sum())
```

```python
#grouping cars based on price range and finding various statistical parameters
data['Price_Range']=pd.cut(data.price,bins=4,labels=['Cheap','Normal','Expensive','Very Expensive'])
print('Car pricing segment and their counts =\n',data.Price_Range.value_counts())
data.groupby(['Price_Range','carbody'])[
    ['wheelbase','carlength','carwidth','curbweight']].agg(['mean','max','min'])
```

# Loan KNN Assignment:

Company wants to automate the loan eligibility process (real time) based on customer detail provided while filling online application form. These details are Gender, Marital Status, Education, Number of Dependents, Income, Loan Amount, Credit History and others. To automate this process, they have given a problem to identify the customers segments, those are eligible for loan amount so that they can specifically target these customers. Here they have provided a partial data set.


```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
loan_data=pd.read_csv(r'/Users/shubhamjadli/Desktop/Term 3/MLP/DataSets/train_u6lujuX_CVtuZ9i (1).csv')

#finding out the total missing rows
print("Missing values in data set = ",loan_data.shape[0]-loan_data.dropna().shape[0])

# Filling in those missing rows based on the most probable 
loan_data.Credit_History.value_counts() # this will be done for all the columsn which have missing data
#Gender most dominant is MALE
#Married most dominant is Yes 
#Dependents most dominant is 0
#Self_Employed most dominant is No
#LoanAmount cannot be used as the values are too close
#Loan_Amount_Term most dominant is 360.0
#Credit_History most dominant is 1.0
loan_data.Gender.fillna('Male',inplace=True)
loan_data.Married.fillna('Yes',inplace=True)
loan_data.Dependents.fillna(0,inplace=True)
loan_data.Self_Employed.fillna('No',inplace=True)
loan_data.Loan_Amount_Term.fillna(360.0,inplace=True)
loan_data.Credit_History.fillna(1.0,inplace=True)

#print(loan_data.isna().sum()) // Here we will only get NA values in LoanAmount
loan_data_final=loan_data.dropna()
print("Total rows/indexes removed after filling in NA values based on most occured values= "
      ,loan_data.shape[0]-loan_data_final.shape[0]) # Only 22 values will get dropped

#finding duplicated data 
print("Duplicated data = ",loan_data_final.duplicated().sum()) 

#Mapping values for categorical
loan_data_final['Gender']=loan_data_final['Gender'].map({'Male':1,'Female':0})
loan_data_final['Married']=loan_data_final.Married.map({'Yes':1,'No':0})
loan_data_final['Dependents']=loan_data_final.Dependents.map({'0':0,'1':1,'3+':3,0:0})
loan_data_final['Education']=loan_data_final.Education.map({'Not Graduate':0,'Graduate':1})
loan_data_final['Property_Area']=loan_data_final.Property_Area.map({'Semiurban':0,'Urban':1,'Rural':2})
loan_data_final['Self_Employed']=loan_data_final.Self_Employed.map({'Yes':1,'No':0})
loan_data_final['Loan_Status']=loan_data_final.Loan_Status.map({'Y':1,'N':0})
loan_data_final.Dependents.fillna(3,inplace=True)

# Bifurcating training and testing data from the orginal data_loan_final
x=loan_data_final.iloc[:,1:12]
y=loan_data_final.iloc[:,12]

#Allocating training and test data using train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=1)

#Passing the training x and y data through the KNN algorithm
KNN=KNeighborsClassifier(n_neighbors=5, metric='euclidean')
KNN.fit(x_train,y_train)

#Predicting the target variable based on some test data 'x_test' 
y_pred=KNN.predict(x_test)
np.array_equal(y_test,y_pred)# checking if all values are same

#Calculating sensitivity, recall and accuracy
Confusion_Matrix=confusion_matrix(y_test, y_pred)
C_M=Confusion_Matrix.flatten(order='F')
print("Precision = ",(C_M[3]/(C_M[2]+C_M[3])*100))
print("Recall/ Sensitivity = ",(C_M[3]/(C_M[0]+C_M[3])*100))
print("Negative predictive value =",(C_M[1]/(C_M[0]+C_M[1])*100))
print("Specificity =",(C_M[1]/(C_M[1]+C_M[2])*100))
print("Accuracy = ",((C_M[3]+C_M[1]))/(C_M[1]+C_M[0]+C_M[2]+C_M[3]))
F_score=(2*(C_M[3]/(C_M[2]+C_M[3])*100)*(C_M[3]/(C_M[0]+C_M[3])*100))/((C_M[3]/(C_M[2]+C_M[3])*100)+(C_M[3]/(C_M[0]+C_M[3])*100))
print("F Score =",F_score)

```

```python
Gen= input("Input Gender \n 1:for Male 0:for Female")
Marr= input("Enter 1 if Married or 0 for Unmarried")
Depen= input("Number of depedents present \n Input 1, 2 or 3+ in case of more than 3")
Edu= input ("Enter education level \n Input 0 for Not Graduate \n 1 for Graduate ")
SelfEmp= input("If Self employed Input 1 or else enter 0")
AppInc= input("Enter Applicant income")
CoApInc=input("Enter Co Applicant income")
LoAmt=input("Enter loan amount")
LoAmtTerm=input("Enter loan amount term")
Crehis=input("Enter credit history")
PropAre=input("Enter where the property area is \n 0 for Semiurban \n 1 for Urban\n 2 for Rural")
x_actual_values=[Gen,Marr,Depen,Edu,SelfEmp,AppInc,CoApInc,LoAmt,LoAmtTerm,Crehis,PropAre]
```
